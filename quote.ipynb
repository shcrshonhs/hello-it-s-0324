## 🔧 2. 도구 불러오기

```python

import requests          # 웹페이지 요청하는 도구
from bs4 import BeautifulSoup  # HTML 분석 도구 (bs4에서 BeautifulSoup 가져옴)
import pandas as pd      # 표 만드는 도구 (pd는 pandas 약자)

```

## 🌐 3. 웹페이지 가져오기

```python

url = 'http://quotes.toscrape.com/'           # 가져올 웹사이트 주소
response = requests.get(url)                  # requests.get() = 지정한 주소의 페이지 요청
soup = BeautifulSoup(response.text, 'html.parser')  # response.text = 가져온 HTML 내용
                                              # BeautifulSoup() = HTML을 분석할 수 있게 정리
                                              # html.parser = HTML 분석 방식 지정
print('페이지 읽기 성공!')                    # print() = 화면에 출력

```

## ✂️ 4. 명언과 작가 추출

```python

quotes = []                                   # quotes = 빈 리스트(명언 저장소) 만들기
for quote_div in soup.find_all('div', class_='quote'):  # soup.find_all() = 모든 quote 클래스 div 찾기
                                                        # for ... in ... = quote_div 하나씩 가져와서 반복
    text = quote_div.find('span', class_='text').text    # .find() = 첫 번째 일치하는 태그 찾기
    # span 태그 중 class='text'인 것의 .text = 텍스트만 뽑기
    author = quote_div.find('small', class_='author').text  # 작가 이름 뽑기 (작은 글씨 small 태그)
    quotes.append({'quote': text, 'author': author})  # append() = 리스트에 항목 추가
                                                      # {'quote': text} = 딕셔너리 (key: value 쌍)

print(f"{len(quotes)}개 명언 추출!")           # f-string = 변수 넣어서 출력 (len() = 리스트 길이)
print(quotes[0])                              # quotes[0] = 리스트 첫 번째 항목

```

## 📊 5. 표 저장

```python
# 5️⃣ 표 만들고 저장
df = pd.DataFrame(quotes)                     # pd.DataFrame() = 리스트를 표로 변환 (df = DataFrame 약자)
df.head()                                     # .head() = 상위 5개 행 출력
df.to_csv('quotes.csv', index=False)          # to_csv() = CSV 파일로 저장
                                              # index=False = 행 번호 저장 안 함
print('CSV 저장 완료! 왼쪽 파일탭에서 다운로드해.')
```
